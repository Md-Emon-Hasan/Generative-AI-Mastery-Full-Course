```python
import os
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

# Define the filename
filename = 'got1.txt'

# Check if the file exists in the current directory
if os.path.isfile(filename):
    # Open and read the file
    with open(filename, 'r', encoding='utf-8') as f:
        corpus = f.read()
        # Tokenize the corpus into sentences
        raw_sentences = sent_tokenize(corpus)
        # Process each sentence and store in the 'story' list
        story = [simple_preprocess(sent) for sent in raw_sentences]
else:
    print(f"The file {filename} does not exist in the current directory.")
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```
